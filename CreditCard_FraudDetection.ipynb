```python
# SPLITTING THE DATASET TO CREATE A ORIGINAL TESTING DATASET FOR FINAL TESTING

from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold  # Ensures that the proportion of each class in the training and testing sets is the same as in the original dataset.

X= df.drop("Class", axis = 1)
y= df['Class']

sss= StratifiedKFold(n_splits = 5, random_state=None, shuffle= False)

for train_index, test_index in sss.split(X,y):
    print("Train:", train_index, "Test:", test_index)
    original_Xtrain, original_Xtest = X.iloc[train_index],X.iloc[test_index]
    original_ytrain, original_ytest = y.iloc[train_index],y.iloc[test_index]
# original_Xtrain, original_Xtest, original_ytrain, original_ytest = train_test_split(X, y, random_state = None)

# Turn into an array
original_Xtrain = original_Xtrain.values
original_Xtest = original_Xtest.values
original_ytrain = original_ytrain.values
original_ytest = original_ytest.values


# See if both the train and test label distribution are similarly distributed
train_unique_label, train_counts_label = np.unique(original_ytrain, return_counts=True)
test_unique_label, test_counts_label = np.unique(original_ytest, return_counts=True)
print("*"*50)
print('Label Distributions: \n')
print(train_counts_label/ len(original_ytrain)*100)
print(test_counts_label/ len(original_ytest)*100)
```

```python
from scipy.stats import norm

fig,ax = plt.subplots(3,3,figsize=(20,10))

list=[3,9,10,12,14,16,17]
for j,i in enumerate(list,start=1):
        r=(j-1)//3
        c=(j-1)%3
        column=f'v{i}_fraud_dist'
        column = new_df[f'V{i}'].loc[new_df["Class"]==1].values
        sns.distplot(column,ax=ax[r][c],fit=norm,color='g')
        ax[r][c].set_title(f'V{i} distribution\nfraud transactions')
plt.tight_layout()
plt.show()
```

```python
# Here V14 is the feature with Gaussian dist. compared to others
# It has highest negative correlation with Class

# We will remove Outliers using IQR

# V14 
v14_fraud =new_df['V14'].loc[new_df['Class']==1].values
q25= np.percentile(v14_fraud,25)
q75= np.percentile(v14_fraud,75)
print('Quartile 25: {} | Quartile 75 {}'.format(q25,q75))

v14_iqr = q75-q25
print(f"IQR is {v14_iqr}")

v14_cutoff = v14_iqr*1.5
v14_lower= q25-v14_cutoff
v14_upper= q75+v14_cutoff

print(f"Cutoff is {v14_cutoff}")
print(f"V14 Lower threshold is {v14_lower}")
print(f"V14 Upper threshold is {v14_upper}")

outliers=[x for x in v14_fraud if x < v14_lower or x > v14_upper]
print(f"Outliers in feature V14 for fraud class are {len(outliers)}")
new_df = new_df.drop(new_df[(new_df['V14']>v14_upper) | (new_df['V14']<v14_lower)].index)
print('----' * 50)

# V12
v12_fraud =new_df['V12'].loc[new_df['Class']==1].values
q25= np.percentile(v12_fraud,25)
q75= np.percentile(v12_fraud,75)
print('Quartile 25: {} | Quartile 75 {}'.format(q25,q75))

v12_iqr = q75-q25
print(f"IQR is {v12_iqr}")

v12_cutoff = v12_iqr*1.5
v12_lower= q25-v12_cutoff
v12_upper= q75+v12_cutoff

print(f"Cutoff is {v12_cutoff}")
print(f"V12 Lower threshold is {v12_lower}")
print(f"V12 Upper threshold is {v12_upper}")

outliers=[x for x in v12_fraud if x < v12_lower or x > v12_upper]
print(f"Outliers in feature V12 for fraud class are {len(outliers)}")
new_df = new_df.drop(new_df[(new_df['V12']>v12_upper) | (new_df['V12']<v12_lower)].index)
print('----' * 50)


# V10
v10_fraud =new_df['V10'].loc[new_df['Class']==1].values
q25= np.percentile(v10_fraud,25)
q75= np.percentile(v10_fraud,75)
print('Quartile 25: {} | Quartile 75 {}'.format(q25,q75))

v10_iqr = q75-q25
print(f"IQR is {v10_iqr}")

v10_cutoff = v10_iqr*1.5
v10_lower= q25-v10_cutoff
v10_upper= q75+v10_cutoff

print(f"Cutoff is {v10_cutoff}")
print(f"V10 Lower threshold is {v10_lower}")
print(f"V10 Upper threshold is {v10_upper}")

outliers=[x for x in v10_fraud if x < v10_lower or x > v10_upper]
print(f"Outliers in feature V10 for fraud class are {len(outliers)}")
new_df = new_df.drop(new_df[(new_df['V10']>v10_upper) | (new_df['V10']<v10_lower)].index)
print('----' * 50)

print("Dataset shape after removal is ",new_df.shape)
```

APPLYING CLASSIFIERS

```python
# Defining and splitting the data, the data is already scaled and balanced

from sklearn.model_selection import train_test_split
X = new_df.drop("Class",axis=1)
Y = new_df["Class"]
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2,random_state=42)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)

```

```python
X_train = X_train.values
X_test = X_test.values
Y_train = Y_train.values
Y_test = Y_test.values
```

```python
# Model implmentation
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

models = {"LogR":LogisticRegression(),
          "KNN": KNeighborsClassifier(),
          "SVM": SVC(),
          "DTC":DecisionTreeClassifier()}
```

```python
for key,classifier in models.items():
    classifier.fit(X_train,Y_train)
    training_score = cross_val_score(classifier, X_train, Y_train,cv = 5)
    print("Classifier",classifier.__class__.__name__," has a training score of ",round(training_score.mean(),2)*100,"% accuracy score")
```

```python
# Using Grid Search CV to find the best parameters
from sklearn.model_selection import GridSearchCV

# Logisitic regression
LogR_params = {"penalty":['l1','l2'],'C':[0.01,0.1,1,10,100,1000]}
grid_LogR = GridSearchCV(LogisticRegression(),LogR_params)
grid_LogR.fit(X_train,Y_train)

best_paramsLR = grid_LogR.best_params_
best_scoreLR = grid_LogR.best_score_
best_estimateLR = grid_LogR.best_estimator_
print("LOGISTIC REGRESSION")
print("Best paramters: ", best_paramsLR)
print("Best score: ", best_scoreLR)
print("Best estimate: ",best_estimateLR)
print("-"*50)

# KNN
KNN_params = {"n_neighbors": [2,3,4],"algorithm":["auto","ball_tree","kd_tree","brute"]}
grid_KNN = GridSearchCV(KNeighborsClassifier(),KNN_params)
grid_KNN.fit(X_train,Y_train)

best_paramsKnn = grid_KNN.best_params_
best_scoreKnn = grid_KNN.best_score_
best_estimateKnn = grid_KNN.best_estimator_
print("KNN")
print("Best paramters: ", best_paramsKnn)
print("Best score: ", best_scoreKnn)
print("Best estimate: ",best_estimateKnn)
print("-"*50)


# SVC
SVC_params = {"C":[0.5,0.7,0.9,1],"kernel":['rbf','poly','sigmoid','linear']}
grid_SVC = GridSearchCV(SVC(),SVC_params)
grid_SVC.fit(X_train,Y_train)

best_paramsSVC = grid_SVC.best_params_
best_scoreSVC = grid_SVC.best_score_
best_estimateSVC = grid_SVC.best_estimator_
print("SVC")
print("Best paramters: ", best_paramsSVC)
print("Best score: ", best_scoreSVC)
print("Best estimate: ",best_estimateSVC)
print("-"*50)

# Decision Tree

Tree_params = {"criterion":['gini','entropy'],"max_depth": [2,3,4],"min_samples_leaf": [5,6,7]}
grid_Tree = GridSearchCV(DecisionTreeClassifier(),Tree_params)
grid_Tree.fit(X_train,Y_train)

best_paramsTree = grid_Tree.best_params_
best_scoreTree = grid_Tree.best_score_
best_estimateTree = grid_Tree.best_estimator_
print("Decision Tree")
print("Best paramters: ", best_paramsTree)
print("Best score: ", best_scoreTree)
print("Best estimate: ",best_estimateTree)
```

```python
import warnings
warnings.filterwarnings('ignore')

LogR_score = cross_val_score(best_estimateLR, X_train, Y_train, cv=5)
print("LogisticRegression Cross Validation Score: ", round(LogR_score.mean()*100,2),"%")

KNN_score = cross_val_score(best_estimateKnn, X_train, Y_train,cv=5)
print("KNN Cross Validation Score: ", round(KNN_score.mean()*100,2),"%")

SVC_score = cross_val_score(best_estimateSVC, X_train, Y_train,cv=5)
print("SVC Cross Validation Score: ", round(SVC_score.mean()*100,2),"%")

Tree_score = cross_val_score(best_estimateTree, X_train, Y_train,cv=5)
print("Decision Tree Cross Validation Score: ", round(Tree_score.mean()*100,2),"%")


```

ROC curve

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_predict

# LogR_pred = best_estimateLR.predict(X_test)
# KNN_pred = best_estimateKnn.predict(X_test)
# SVC_pred = best_estimateSVC.predict(X_test)
# Tree_pred = best_estimateTree.predict(X_test)

# print("ROC AUC Score")
# print("Logistic regression - ", roc_auc_score(Y_test,LogR_pred))
# print("KNN - ",roc_auc_score(Y_test,KNN_pred))
# print("SVC - ",roc_auc_score(Y_test,SVC_pred))
# print("Decision Tree - ",roc_auc_score(Y_test,Tree_pred))

LogR_pred = cross_val_predict(best_estimateLR,X_train,Y_train,cv=5,method='decision_function')
KNN_pred = cross_val_predict(best_estimateKnn,X_train,Y_train,cv=5)
SVC_pred = cross_val_predict(best_estimateSVC,X_train,Y_train,cv=5,method='decision_function')
Tree_pred = cross_val_predict(best_estimateTree,X_train,Y_train,cv=5)

print("ROC AUC Score")
print("Logistic regression - ", roc_auc_score(Y_train,LogR_pred))
print("KNN - ",roc_auc_score(Y_train,KNN_pred))
print("SVC - ",roc_auc_score(Y_train,SVC_pred))
print("Decision Tree - ",roc_auc_score(Y_train,Tree_pred))
```

```python
log_fpr,log_tpr, log_threshold = roc_curve(Y_train,LogR_pred)
knn_fpr, knn_tpr, knn_threshold = roc_curve(Y_train,KNN_pred)
svc_fpr,svc_tpr,svc_threshold = roc_curve(Y_train,SVC_pred)
tree_fpr,tree_tpr,tree_threshold = roc_curve(Y_train,Tree_pred)

def plot_roc_curve(log_fpr,log_tpr,knn_fpr, knn_tpr, svc_fpr, svc_tpr, tree_fpr, tree_tpr):
    plt.figure(figsize=(16,8))
    plt.title(" ROC Curve of Classifiers", fontsize=20)
    plt.plot(log_fpr,log_tpr, label=f"Logistic regression Score - {round(roc_auc_score(Y_train,LogR_pred),3)}")
    plt.plot(knn_fpr,knn_tpr, label=f"KNN Score - {round(roc_auc_score(Y_train,KNN_pred),3)}")
    plt.plot(svc_fpr,svc_tpr,label=f'SVC Score - {round(roc_auc_score(Y_train,SVC_pred),3)}')
    plt.plot(tree_fpr,tree_tpr, label=f"Decision tree Score - {round(roc_auc_score(Y_train,Tree_pred),3)}")

    plt.xlabel("False positive rate",fontsize=16)
    plt.ylabel("True positive rate",fontsize=16)
    plt.legend()

plot_roc_curve(log_fpr,log_tpr,knn_fpr,knn_tpr,svc_fpr,svc_tpr,tree_fpr,tree_tpr)
plt.show()
```
